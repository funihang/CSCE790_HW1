```python
import numpy as np

##%matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
## plot function

def plotSurface(x1, x2, y, title):   
    # meshed point: x1, x2
    # function output: y
    # title: figure title

    fig = plt.figure()
    ax = plt.axes(projection='3d')
    
    ## plot surface
    surf = ax.plot_surface(x1, x2, y, cmap='rainbow')
    #surf = ax.plot_trisurf(x1, x2, y, cmap=cm.jet, linewidth=0)
    #ax.scatter(x1, x2, y, color='k')
    
    plt.title(title)
    plt.show()
```

```python
def sigmoid(X):
    y = 1/(1+np.exp(-X))
    return y


def hardLimit(X):
    if X >= 0:
        y = 1
    else:
        y = 0
    return y

def radialBasis(X):  
    # set mu as 0
    # set sigma as 0.5
    return np.exp(-(X ** 2) / (2 * 0.5 ** 2))


def yFunc(act_func, x1, x2):
    
    #print(x1.shape, x2.shape)
    x = np.vstack((x1.flatten(), x2.flatten()))
    
    ## v_T, b_v, b_w, and W
    V_t = np.array([[-2.69, -2.8], [-3.39, -4.56]])
    b_v = np.array([[-2.21], [4.76]])
    b_w = -2.28
    W = np.array([[-4.91], [4.95]])
    
    ## V_T * x + b_v
    X = np.dot(V_t, x) + b_v
    
    ## activation function
    if act_func == 'sigmoid':
        act_res = sigmoid(X)
        
    elif act_func == 'hardLimit':
        act_res = np.vectorize(hardLimit)(X)
        
    elif act_func == 'RBF':
        act_res = np.vectorize(radialBasis)(X)
        
    #print(act_res.shape)    
    
    ## y function
    y = np.dot(W.T, act_res) + b_w
    
    return y
```

## Sigmoid

```python
# x1, x2 100 samples
x1 = np.linspace(-2, 2, 100)
x2 = np.linspace(-2, 2, 100)
x1, x2 = np.meshgrid(x1, x2)

y = yFunc('sigmoid', x1, x2).reshape(x1.shape)
#print(y.shape)
plotSurface(x1, x2, y, 'sigmoid: 100 samples')



# x1, x2 500 samples
x1 = np.linspace(-2, 2, 500)
x2 = np.linspace(-2, 2, 500)
x1, x2 = np.meshgrid(x1, x2)

y = yFunc('sigmoid', x1, x2).reshape(x1.shape)
#print(y.shape)
plotSurface(x1, x2, y, 'sigmoid: 500 samples')



# x1, x2 10000 samples
x1 = np.linspace(-2, 2, 10000)
x2 = np.linspace(-2, 2, 10000)
x1, x2 = np.meshgrid(x1, x2)

y = yFunc('sigmoid', x1, x2).reshape(x1.shape)
#print(y.shape)
plotSurface(x1, x2, y, 'sigmoid: 10000 samples')
```
