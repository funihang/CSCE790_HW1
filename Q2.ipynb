```python
import numpy as np

#%matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
## plot function

def plotSurface(x1, x2, y, title):   
    # meshed point: x1, x2
    # function output: y
    # title: figure title

    fig = plt.figure()
    ax = plt.axes(projection='3d')
    
    
    surf = ax.plot_surface(x1, x2, y, cmap='rainbow')
    #surf = ax.plot_trisurf(x1, x2, y, cmap=cm.jet, linewidth=0)
    #ax.scatter(x1, x2, y, color='k')
    
    plt.title(title)
    plt.show()
```

## sigmoid

```python
def sigmoid(X):
    y = 1/(1+np.exp(-X))
    
    return y


# x1, x2 100 samples
x1 = np.linspace(-2, 2, 100)
x2 = np.linspace(-2, 2, 100)
x1, x2 = np.meshgrid(x1, x2)
# print(x1.shape, x2.shape)

# when the activation function is sigmoid
y = sigmoid(-0.479*x1 + 5.90*x2-0.93)
# print(y.shape, type(y), y)
plotSurface(x1, x2, y,'sigmoid: 100 samples')


# x1, x2 5000 samples
x1 = np.linspace(-2, 2, 5000)
x2 = np.linspace(-2, 2, 5000)
x1, x2 = np.meshgrid(x1, x2)

# when the activation function is sigmoid
y = sigmoid(-0.479*x1 + 5.90*x2-0.93)
plotSurface(x1, x2, y, 'sigmoid: 5000 samples')


# x1, x2 10000 samples
x1 = np.linspace(-2, 2, 10000)
x2 = np.linspace(-2, 2, 10000)
x1, x2 = np.meshgrid(x1, x2)

# when the activation function is sigmoid
y = sigmoid(-0.479*x1 + 5.90*x2-0.93)
print(y.shape)
plotSurface(x1,x2,y, 'sigmoid: 10000 samples')
```

## Hard limit

```python
def hardLimit(X):
    if X >= 0:
        y = 1
    else:
        y = 0
    return y


# x1, x2 100 samples
x1 = np.linspace(-2, 2, 100)
x2 = np.linspace(-2, 2, 100)
x1, x2 = np.meshgrid(x1, x2)

# when the activation function is sigmoid
#y = hardLimit((-0.479*x1 + 5.90*x2-0.93).any())
#print(y)
y = np.vectorize(hardLimit)(-0.479*x1 + 5.90*x2-0.93)
print(y.shape)

plotSurface(x1,x2,y, 'hard limit: 100 samples')


# x1, x2 5000 samples
x1 = np.linspace(-2, 2, 5000)
x2 = np.linspace(-2, 2, 5000)
x1, x2 = np.meshgrid(x1, x2)

y = np.vectorize(hardLimit)(-0.479*x1 + 5.90*x2-0.93)
print(y.shape)

plotSurface(x1,x2,y, 'hard limit: 5000 samples')


# x1, x2 10000 samples
x1 = np.linspace(-2, 2, 10000)
x2 = np.linspace(-2, 2, 10000)
x1, x2 = np.meshgrid(x1, x2)

y = np.vectorize(hardLimit)(-0.479*x1 + 5.90*x2-0.93)
print(y.shape)

plotSurface(x1,x2,y, 'hard limit: 10000 samples')
```

## Radial basis function

```python
def radialBasis(X):  
    ## set mu as 0
    ## set sigma as 0.5
    return np.exp(-(X ** 2) / (2 * 0.5 ** 2))

# 100 samples
# x1, x2 domain
x1 = np.linspace(-2, 2, 100)
x2 = np.linspace(-2, 2, 100)
x1, x2 = np.meshgrid(x1, x2)

# when the activation function is sigmoid
y = np.vectorize(radialBasis)(-0.479*x1 + 5.90*x2-0.93)

plotSurface(x1,x2,y, 'RBF: 100 samples')


# 500 samples
# x1, x2 domain
x1 = np.linspace(-2, 2, 500)
x2 = np.linspace(-2, 2, 500)
x1, x2 = np.meshgrid(x1, x2)

# when the activation function is sigmoid
y = np.vectorize(radialBasis)(-0.479*x1 + 5.90*x2-0.93)

plotSurface(x1,x2,y, 'RBF: 500 samples')


# 10000 samples
# x1, x2 domain
x1 = np.linspace(-2, 2, 10000)
x2 = np.linspace(-2, 2, 10000)
x1, x2 = np.meshgrid(x1, x2)

# when the activation function is sigmoid
y = np.vectorize(radialBasis)(-0.479*x1 + 5.90*x2-0.93)

plotSurface(x1,x2,y, 'RBF: 10000 samples')
```
